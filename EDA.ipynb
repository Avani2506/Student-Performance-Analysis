```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cleaned_dataset.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df['final_exam_score'].hist()
plt.title("Distribution")
plt.show()
```

```python
sns.boxplot(x=df['final_exam_score'])
plt.show()
```

```python
df['grade'].hist()
plt.title("Distribution")
plt.show()
```

```python
sns.boxplot(x=df['grade'])
plt.show()
```

```python
plt.scatter(df['midterm_score'], df['overall_score'])
plt.xlabel("midterm_score")
plt.ylabel("overall_score")
plt.show()
```

```python
sns.boxplot(x='gender', y='overall_score', data=df)
plt.show()
```

```python
corr = df.select_dtypes(include=[np.number]).corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

```python
df['gender'] = df['gender'].map({'Male': 0, 'Female': 1})
print(df['gender'].value_counts())
```

```python
df['internet_access'] = df['internet_access'].map({'No': 0, 'Yes': 1})
print(df['internet_access'].value_counts())
```

```python
df['extra_classes'] = df['extra_classes'].map({'No': 0, 'Yes': 1})
print(df['extra_classes'].value_counts())
```

```python
edu_order = {'High School': 1, 'Bachelor': 2, 'Master': 3, 'PhD': 4}
df['parent_education'] = df['parent_education'].map(edu_order)
print(df['parent_education'].value_counts())
```

```python
grade_order = {'F': 0, 'D': 1, 'C': 2, 'B': 3, 'A': 4}
df['grade'] = df['grade'].map(grade_order)
print(df['grade'].value_counts())
```

```python
print(df.dtypes)
print(df.head())
```

```python
df.mean()
df.median()
df.std()
```

```python
import statsmodels.api as sm

sm.qqplot(df['grade'], line='45')
plt.show()
```
